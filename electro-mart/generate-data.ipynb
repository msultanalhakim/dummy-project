```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(42)

n_rows = 10000

# Data Dasar (Master Data Produk)
products = {
    'Laptop Gaming': 15000000,
    'Mouse Wireless': 150000,
    'Monitor 24in': 2000000,
    'Keyboard Mech': 800000,
    'Headset RGB': 500000,
    'Kabel HDMI': 50000 # Murah
}
cities = ['Jakarta', 'Surabaya', 'Bandung', 'Medan', 'Makassar', 'jakarta ', 'SURABAYA'] # Perhatikan typo/spasi

# Generate Kolom Transaksi
data = {
    'order_id': np.arange(1, n_rows + 1),
    'product_name': np.random.choice(list(products.keys()), n_rows),
    'city': np.random.choice(cities, n_rows),
    'quantity': np.random.randint(1, 5, n_rows), # Beli 1-4 barang
    'order_date': [datetime(2024, 1, 1) + timedelta(days=np.random.randint(0, 365)) for _ in range(n_rows)],
}

# Buat DataFrame
df = pd.DataFrame(data)

# Tambahkan Harga (Mapping dari dictionary products)
df['price_per_unit'] = df['product_name'].map(products)

# Mengotori Data

# Masukkan Missing Values (NaN)
# Ada 5% transaksi yang quantity-nya tidak tercatat (Error sistem)
df.loc[df.sample(frac=0.05).index, 'quantity'] = np.nan 

# Masukkan Tanggal Pengiriman (Shipping Date)
# Normalnya kirim 1-3 hari, tapi ada yang telat sampai 10 hari
df['shipping_date'] = df['order_date'] + pd.to_timedelta(np.random.randint(1, 10, n_rows), unit='D')

# Bikin Error di Tanggal Pengiriman (Ada yang belum dikirim / NULL / NaT)
df.loc[df.sample(frac=0.1).index, 'shipping_date'] = pd.NaT

# Ubah data order_date jadi String
df['order_date'] = df['order_date'].astype(str)

# Save ke CSV
df.to_csv('electromart.csv', index=False)

print(f"File 'electromart.csv' dengan {n_rows} baris berhasil dibuat!")
```
